```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("seaborn")

%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

#Univarient analysis

```python
sns.distplot(tips["total_bill"])
```

```python
sns.displot(tips["total_bill"])
```

```python
sns.displot(tips["total_bill"], kde = True)
```

```python
tips.shape
```

```python
sns.displot(tips["total_bill"], kde = True, bins = 10)
```

```python
sns.displot(tips["tip"], kde = True, bins = 10)
```

# Bivarient Analysis

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
```

# Plotly

```python
import plotly.express as px
```

```python
x = [10, 20, 30, 40]
y = [10, 20, 7, 23]

px.line(x = x, y = y)
```

```python
x = [10, 20, 30, 40]
y = [10, 20, 7, 23]
z = [23, 56, 24, 45]
px.line_3d(x = x, y = y, z = z)
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
fig = px.scatter(iris, x = "sepal_length", y = "sepal_width", color = "species", size="petal_length")
fig.show()
```

```python
fig = px.scatter_3d(iris, x = "sepal_length", y = "sepal_width", color = "species", z="petal_length")
fig.show()
```

```python
px.bar(iris, x = "sepal_length", y = "sepal_width")
```

```python
px.pie(tips, names = 'day', values = 'tip')
```

```python
import plotly.graph_objects as go
import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=go.Surface(z=z_data))

fig.show()
```
