# Transformers

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer


cat_col = ['ocean_proximity']
num_col = list(df.drop([*cat_col, 'median_house_value'], axis = 1).columns)

col_transf = ColumnTransformer([
                                ('onehot_cat', OneHotEncoder(), cat_col),
                                ('inputer', SimpleImputer(strategy = "median"), num_col),
                                ('num_scaling1', StandardScaler(), num_col),
])
```

# Select and Train Model

```python
Y = df['median_house_value']
X = df.drop('median_house_value', axis = 1)
```

```python
X = col_transf.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(X, Y)
```

```python
y_pred = lin.predict(X)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y, y_pred)
mse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor()
tree.fit(X, Y)
```

```python
y_pred = tree.predict(X)

mse = mean_squared_error(Y, y_pred)
mse
```

```python
y_pred
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

xtrain, xval, ytrain, yval = train_test_split(X, Y, train_size=0.7)
```

```python
xtrain.shape, xval.shape, ytrain.shape, yval.shape
```

```python
lin.fit(xtrain, ytrain)
tree.fit(xtrain, ytrain)

lin_predict = lin.predict(xval)
tree_predict = tree.predict(xval)
```

```python
print('Linear Regression: ', mean_squared_error(lin_predict, yval))
print('Decision Tree    : ', mean_squared_error(tree_predict, yval))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(xtrain, ytrain)
forest_pred = forest.predict(xval)

print('Random Forest    : ', mean_squared_error(forest_pred, yval))
```

```python
print('Linear Regression: ', mean_squared_error(lin_predict, yval))
print('Decision Tree    : ', mean_squared_error(tree_predict, yval))
print('Random Forest    : ', mean_squared_error(forest_pred, yval))
```
