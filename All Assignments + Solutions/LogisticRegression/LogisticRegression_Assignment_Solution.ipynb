#Feature Engineering

```python
def set_bmi(row):
    if row < 18.5:
        return "Under"
    elif row <= 24.9:
        return "Healthy"
    elif row <= 29.9:
        return "Over"
    elif row >= 30:
        return "Obese"

df['BMI_STAT'] = df["BMI"].apply(set_bmi)
```

```python
sns.countplot(data=df, x='BMI_STAT', hue='Outcome')
```

```python
def set_insulin(row):
    if row >= 16 and row <= 166:
        return "Normal"
    else:
        return "Abnormal"

df['Insulin_STAT'] = df['Insulin'].apply(set_insulin)

sns.countplot(data=df, x='Insulin_STAT', hue='Outcome')
```

```python
df.corr()['Outcome'].sort_values(ascending=False)
```

```python
df['Pregnancies_STAT'] = np.where(df['Pregnancies'] <= 6,0,1)
```

```python
df.columns
```

```python
df1 = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']]

df2 = df[['Glucose', 'BloodPressure', 'SkinThickness',
           'DiabetesPedigreeFunction', 'Age', 'Outcome',
          'BMI_STAT', 'Insulin_STAT', 'Pregnancies']]
```

```python
df2 = df2.join(pd.get_dummies(df2[['BMI_STAT', 'Insulin_STAT']], drop_first=True))
df2.drop(['BMI_STAT', 'Insulin_STAT'], axis = 1, inplace = True)
df2.sample(4)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
df_train = df2
X = df_train.drop(['Outcome'], axis=1)
y = df_train['Outcome']

X = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split 
xtrain, xtest , ytrain , ytest = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(xtrain, ytrain)
```

```python
log.score(xtest, ytest)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': np.linspace(0.0, 1.5, 20), 'penalty':['l1', 'l2','elastic','none']}
grid = GridSearchCV(log, param_grid, cv=10)
grid.fit(xtrain, ytrain)
```

```python
grid.best_score_, grid.best_params_
```
