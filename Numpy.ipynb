```python
l = [1,2,3,4]
l2 = l[1:3]
print(l)
print(l2)
l2[0] = 1000
print(l)
print(l2)
```

```python
arr = np.array([1,2,3,4])
arr2 = arr[1:3]
print(arr)
print(arr2)
arr2[0]= 1000
print(arr)
print(arr2)
```

```python
arr = np.array([1,2,3,4])
arr2 = arr[1:3].copy()
print(arr)
print(arr2)
arr2[0] = 1000
print(arr)
print(arr2)
```
