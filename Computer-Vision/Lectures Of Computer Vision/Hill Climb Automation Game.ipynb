```python

mp_drawing = mp.solutions.drawing_utils
holistic =mp.solutions.holistic
mp_pose = mp.solutions.pose
```

```python
cap = cv2.VideoCapture(0)
with holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holis :
    while 1: 
        _,frame = cap.read()
        fr = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        frame = cv2.flip(frame,1)
#         fr = cv2.flip(fr,1)
        results = holis.process(fr)
        height,width,_ = frame.shape
        print(frame.shape)
        pose="move"
        y_mid = height//2
        if results.right_hand_landmarks  :
            right_hand = (results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x * width,
                         results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y * height)
           
            x=right_hand[0]
            pose= "MOVE"
            if(right_hand[1]<y_mid):
                pose="acc"
                pydirectinput.keyUp('left')
                pydirectinput.keyDown('right')
            elif(right_hand[1]>y_mid):
                pose="break"
                pydirectinput.keyUp('right')
                pydirectinput.keyDown('left')

            else:
                pose="Move"
                pydirectinput.keyUp('right')
                pydirectinput.keyUp('left')
        cv2.line(frame,(0,y),(width,y),(0,0,255),2)
        cv2.putText(frame,pose,(30,30),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),2)
        cv2.imshow('win1',frame)
        if cv2.waitKey(1)==ord('q'):
                break;
cap.release()
cv2.destroyAllWindows()
```
