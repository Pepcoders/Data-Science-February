# Bar Chart

```python
country = ["India", "US", "UK", "Ukraine", "Russia"]
male = [87, 45, 76, 23, 35]
female = [12, 32, 25, 32, 21]
```

```python
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
plt.bar(country, male)
plt.xlabel("Country")
plt.ylabel("Male")
plt.title("Populations")
```

```python
plt.bar(country, male, width = 0.4, color="orange")
plt.xlabel("Country")
plt.ylabel("Male")
plt.title("Populations")
```

```python
plt.bar(country, male)
plt.bar(country, female, width = 0.4)
plt.xlabel("Country")
plt.ylabel("Male")
plt.title("Populations")
```

```python
plt.bar(country, male, width = 0.5)
plt.bar(country, female, width = 0.5,align = 'edge')
plt.xlabel("Country")
plt.ylabel("Male")
plt.title("Populations")
```

```python
plt.bar(country, male, label = "Male")
plt.bar(country, female, bottom = male, label = "Female")
plt.xlabel("Country")
plt.ylabel("Male")
plt.title("Populations")
plt.legend(shadow = True, facecolor = "white", frameon = True)
```

# Histogram

```python
import numpy as np
np.random.seed(42)

num = np.random.randn(100)
```

```python
plt.hist(num, color = 'orange')
```

```python
plt.hist(num, color = 'orange', bins = 10)
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/purna135/ML-Datasets/main/iris.csv")
df.head()
```

```python
plt.hist(df["sepal.length"], alpha = 0.5, label = "Sepal length")
plt.hist(df["petal.length"], alpha = 0.5, label = "Petal length")
plt.legend(shadow = True, facecolor = "white", frameon = True)
```

# Scatter

```python
plt.scatter(df["sepal.length"], df["sepal.width"], marker = "*")
plt.scatter(df["petal.length"], df["petal.width"], marker = "d")
```

# Pie

```python
plt.pie(male, labels = country)
plt.show()
```

```python
plt.pie(male, labels = country, autopct = "%1.1f%%", shadow=True, explode = [0.2, 0, 0, 0, 0])
plt.show()
```

# Subplot

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df["sepal.width"])

plt.subplot(1, 2, 2)
plt.hist(df["petal.width"])
```
