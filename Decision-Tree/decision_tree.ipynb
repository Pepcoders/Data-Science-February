```python
x = np.arange(6)

y= x%2
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,c=y)
```

```python
X = x.reshape(-1,1)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(X,y)
```

```python
model.score(X,y)
```

```python
from sklearn import tree 
```

```python
plt.figure(figsize = (10,10))
tree.plot_tree(model,filled = True)
plt.show()
```

```python
import numpy as np
```

```python
p1= np.linspace(0,1,15)
p2 =1-p1
ye = -p1*np.log2(p1)-p2*np.log2(p2)
yg = 1 - p1*p1-p2*p2
ye[0] =0
ye[-1] =0
```

```python
plt.plot(p1,ye,label='entropy')
plt.plot(p1,yg,label='ginny')
plt.legend()
```
