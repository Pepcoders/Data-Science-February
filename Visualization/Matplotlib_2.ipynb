# Bar Chart

```python
country = ["India", "US", "UK", "Russia", "Ukraine"]
male = [79, 32, 67, 82, 23, ]
female = [32, 34, 23, 34, 21]
```

```python
plt.bar(country, male)
plt.xlabel("Country")
plt.ylabel("Male population")
plt.title("Population")
```

```python
plt.bar(country, male, color = "Orange", width = 0.5)
plt.xlabel("Country")
plt.ylabel("Male population")
plt.title("Population")
```

```python
plt.bar(country, male, color = "Orange", width = 0.5)
plt.bar(country, female, color = "DarkTurquoise", width = 0.5, align = "edge")
plt.xlabel("Country")
plt.ylabel("Male population")
plt.title("Population")
```

```python
plt.bar(country, male, color = "Orange", width = 0.5, label = "Male")
plt.bar(country, female, color = "DarkTurquoise", bottom = male, width = 0.5, label = "Female")
plt.xlabel("Country")
plt.ylabel("Male population")
plt.title("Population")

plt.legend()
```

```python
plt.bar(country, male, color = "Orange", width = 0.5, label = "Male")
plt.bar(country, female, color = "DarkTurquoise", bottom = male, width = 0.5, label = "Female")
plt.xlabel("Country")
plt.ylabel("Male population")
plt.title("Population")

plt.legend(shadow = True, facecolor = "white", frameon = True)
```

# Histogram

```python
import numpy as np
np.random.seed(42)

num = np.random.randn(100)
plt.hist(num)
plt.show()
```

```python
plt.hist(num, color = "SandyBrown")
plt.show()
```

```python
plt.hist(num, color = "SandyBrown", bins = 20)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/purna135/ML-Datasets/main/iris.csv")
df.head()
```

```python
df["sepal.length"]
```

```python
plt.hist(df["sepal.length"], alpha = 0.5, label = "Sepal Length")
plt.hist(df["petal.length"], alpha = 0.5, label = "Petal Length")
plt.hist(df["petal.width"], alpha = 0.5, label = "Petal Width")
plt.legend(frameon = True, facecolor = "white", shadow = True)
```

# Scatter 

```python
plt.scatter(df["sepal.length"], df["sepal.width"], marker = 'd', label = "Sepal Length")
plt.scatter(df["petal.length"], df["petal.width"], marker = 'o', label = "Petal Length")
plt.legend()
```

```python
np.random.seed(10)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
